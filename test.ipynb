```python
import diyplot as diyp
import numpy as np
from matplotlib import pyplot as plt
from cmcrameri import cm

x = np.random.uniform(0, 1, size=(512, 512))
fig, ax = plt.subplots(ncols=3, figsize=(10, 3))

(fig, _) = diyp.imshow_with_cbar(fig, ax[0], x, cmap_type='cyclic', title_str='cyclic')
(fig, _) = diyp.imshow_with_cbar(fig, ax[1], x, cmap_type='sequential', title_str='sequential')
(fig, _) = diyp.imshow_with_cbar(fig, ax[2], x, cmap_type='diverging', title_str='diverging')
```
