```python
import diyplot as diyp
import numpy as np
from matplotlib import pyplot as plt
from cmcrameri import cm
from scipy.misc import ascent, face
face = face(gray=True)
ascent = ascent()

fig, ax = plt.subplots(ncols=4, figsize=(13, 3))

(fig, _) = diyp.imshow_with_cbar(ascent, fig, ax[0], cmap_type='cyclic', title_str='cyclic')
(fig, _) = diyp.imshow_with_cbar(ascent, fig, ax[1], cmap_type='sequential', title_str='sequential')
(fig, _) = diyp.imshow_with_cbar(ascent, fig, ax[2], cmap_type='diverging', title_str='diverging')
(fig, _) = diyp.imshow_with_cbar(ascent, fig, ax[3], cmap_type='batlow', title_str='batlow')
```
